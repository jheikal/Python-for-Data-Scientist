```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#df = pd.read_excel('http://archive.ics.uci.edu/ml/machine-learning-databases/00352/Online%20Retail.xlsx')
df = pd.read_excel('Online Retail.xlsx')
df.head()
```

```python
df.shape
```

```python
# Clean up spaces in description and remove any rows that don't have a valid invoice
df['Description'] = df['Description'].str.strip()
```

```python
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
```

```python
df['InvoiceNo'] = df['InvoiceNo'].astype('str')
df = df[~df['InvoiceNo'].str.contains('C')]
```

```python
df['InvoiceNo'].value_counts().count()
```

```python
df.shape
```

```python
df['Country'].value_counts()
```

```python
df['Country'].value_counts().count()
```

```python
len(df['Country'].value_counts())
```

```python
df[df['Country'] =="Bahrain"].head(20)
```

```python
basket = (df[df['Country'] =="Bahrain"]
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('InvoiceNo'))
```

```python
basket.head()
```

```python
basket.shape
```

```python
# Show a subset of columns
basket.iloc[:,[0,1,2,3,4,5,6, 7]].head()
```

```python
# Convert the units to 1 hot encoded values
def encode_units(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
```

```python
basket.columns
```

```python
basket_sets = basket.applymap(encode_units)
```

```python
# No need to track postage, only if region is country
# basket_sets.drop('POSTAGE', inplace=True, axis=1)
```

```python
basket_sets.head()
```

```python
basket_sets.shape
```

```python
# Build up the frequent items# Build  
frequent_itemsets = apriori(basket_sets, min_support=0.3, use_colnames=True)
```

```python
frequent_itemsets.shape
```

```python
frequent_itemsets.head()
```

```python
frequent_itemsets.shape
```

```python
# Create the rules# Create 
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules
```

```python
rules.shape
```

```python
rules[ (rules['lift'] >= 3) &
       (rules['confidence'] >= 0.8) ]
```
